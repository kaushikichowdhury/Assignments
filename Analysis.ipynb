```python
# Initialize list to store data
data = []

# Function to extract content from a given URL
def scrape_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        # Extract title
        title = soup.find("h1").text.strip() if soup.find("h1") else "No title found"

        # Try different containers for article content
        content_classes = [r"^Art-exp_\w+", "_s30J clearfix"]

        content = ""
        for class_name in content_classes:
            paragraphs = soup.find_all("div", {"class": class_name}) or soup.find_all("p", {"class": class_name})
            if paragraphs:
                content = " ".join([para.text.strip() for para in paragraphs])
                break

        # If content is still empty, search for <p> tags directly
        if not content:
            paragraphs = soup.find_all("p")
            content = " ".join([para.text.strip() for para in paragraphs if para.text.strip()])

        # Return dictionary with title, content, and URL
        return {"URL": url, "Title": title, "Content": content if content else "No content found"}

    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return {"URL": url, "Title": "Error", "Content": "Error"}

# Loop through URLs and scrape content
for url in urls:
    data.append(scrape_content(url))

# Save data to a CSV file
df = pd.DataFrame(data)
df.to_csv("scraped_articles.csv", index=False, encoding="utf-8")

print("✅ Content scraped and saved to 'scraped_articles.csv'")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# Initialize list to store data
data = []

# Function to extract content and metadata from a given URL
def scrape_content(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        # Extract title
        title = soup.find("h1").text.strip() if soup.find("h1") else "No title found"

        # Extract metadata
        meta_description = soup.find("meta", attrs={"name": "description"})
        meta_keywords = soup.find("meta", attrs={"name": "keywords"})
        pub_date = soup.find("meta", attrs={"property": "article:published_time"}) or \
                   soup.find("meta", attrs={"name": "pubdate"}) or \
                   soup.find("meta", attrs={"itemprop": "datePublished"})
        author = soup.find("meta", attrs={"name": "author"}) or soup.find("span", class_=re.compile("author"))

        # Clean and assign metadata
        meta_description = meta_description['content'].strip() if meta_description and meta_description.has_attr('content') else "N/A"
        meta_keywords = meta_keywords['content'].strip() if meta_keywords and meta_keywords.has_attr('content') else "N/A"
        pub_date = pub_date['content'].strip() if pub_date and pub_date.has_attr('content') else "N/A"
        author = author['content'].strip() if author and author.has_attr('content') else (author.text.strip() if author else "N/A")

        # Try different containers for article content
        content_classes = [r"^Art-exp_\w+", "_s30J clearfix"]
        content = ""

        for class_name in content_classes:
            paragraphs = soup.find_all("div", {"class": re.compile(class_name)}) or soup.find_all("p", {"class": re.compile(class_name)})
            if paragraphs:
                content = " ".join([para.text.strip() for para in paragraphs])
                break

        # If content is still empty, search for <p> tags directly
        if not content:
            paragraphs = soup.find_all("p")
            content = " ".join([para.text.strip() for para in paragraphs if para.text.strip()])

        # Return all extracted data
        return {
            "URL": url,
            "Title": title,
            "Author": author,
            "Publication Date": pub_date,
            "Meta Description": meta_description,
            "Meta Keywords": meta_keywords,
            "Content": content if content else "No content found"
        }

    except Exception as e:
        print(f"❌ Failed to scrape {url}: {e}")
        return {
            "URL": url,
            "Title": "Error",
            "Author": "Error",
            "Publication Date": "Error",
            "Meta Description": "Error",
            "Meta Keywords": "Error",
            "Content": "Error"
        }

# Loop through URLs and scrape content + metadata
for url in urls:
    data.append(scrape_content(url))

# Save data to a CSV file
df = pd.DataFrame(data)
df.to_csv("scraped_articles_with_metadata.csv", index=False, encoding="utf-8")

print("✅ Content and metadata scraped and saved to 'scraped_articles_with_metadata.csv'")
```

## NER
This section employs Named Entity Recognition (NER) to extract and analyze the 20 most frequently mentioned geographical locations within the collected text corpus. The ranked locations allow for a targeted understanding of where such cases are most prevalent, providing valuable insights.

## Bigram Analysis
Bigram analysis was employed to identify and examine the most frequently occurring pairs of consecutive words within the dataset. This technique aimed to uncover commonly used phrases and recurring themes in the reportage of dog cruelty cases.

Here too, normalisation was done according to authorial discretion to highlight only the unique instances.

```python
# Install required packages (if not already installed)
!pip install nltk
!pip install pandas
!pip install matplotlib
!pip install wordcloud

# Import libraries
import pandas as pd
import nltk
from nltk.util import ngrams
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import string
```

```python
nltk.download('all')
```

```python
# === 1. Load CSV from existing file ===
file_path = '/content/scraped_articles_with_metadata copy.csv'  # <-- Replace with your actual filename
column_name = 'Content'  # <-- Replace with the column name containing text

df = pd.read_csv(file_path)
texts = df[column_name].dropna().astype(str)

# === 2. Preprocessing ===
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return tokens
```

```python
from nltk.stem import WordNetLemmatizer
# === 3. Normalization ===
lemmatizer = WordNetLemmatizer()
custom_stopwords = set(['animal', 'animals', 'cruelty'])

def normalize_bigram(bigram):
    w1, w2 = [lemmatizer.lemmatize(word) for word in bigram]

    # Skip bigram if either word is in custom stopwords
    if w1 in custom_stopwords or w2 in custom_stopwords:
        return None

    return (w1, w2)
```

```python
# === 4. Semantic Mapping & Normalization ===
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
custom_stopwords = set(['animal', 'animals', 'cruelty'])

# Semantic mapping for equivalent bigrams (you can expand this)
equivalent_bigrams = {
    ('case', 'registered'): ('police', 'complaint'),
    ('police', 'registered'): ('police', 'complaint'),
    ('fir', 'registered'): ('police', 'complaint'),
    ('police', 'said'): ('police', 'complaint'),
    ('police', 'station'): ('police', 'complaint'),
    ('registered', 'fir'): ('police', 'complaint'),

    ('indian', 'penal'): ('penal', 'code'),

    ('social', 'medium'): ('social', 'media'),

    ('street', 'dog'): ('stray', 'dog'),
    ('stray', 'dog'): ('stray', 'dog'),
    ('stray', 'dogs'): ('stray', 'dog')
}

def normalize_bigram(bigram):
    w1, w2 = [lemmatizer.lemmatize(word.lower()) for word in bigram]

    # Filter out if either word is in the custom stopword list
    if w1 in custom_stopwords or w2 in custom_stopwords:
        return None

    # Return mapped semantic group if available
    key = (w1, w2)
    if key in equivalent_bigrams:
        return equivalent_bigrams[key]

    return (w1, w2)
```

```python
# === 5. Generate and normalize bigrams ===
all_bigrams = []
for text in texts:
    tokens = preprocess(text)
    bigrams = list(ngrams(tokens, 2))
    for bigram in bigrams:
        normalized = normalize_bigram(bigram)
        if normalized:
            all_bigrams.append(normalized)

# === 6. Count normalized bigrams ===
bigram_freq = Counter(all_bigrams)
common_bigrams = bigram_freq.most_common(20)

# === 7. Print top bigrams ===
print("Top 20 Bigrams (Semantic Normalized):")
for bigram, freq in common_bigrams:
    print(f"{bigram}: {freq}")
```

```python
# === 5. Visualization (Bar Chart) ===
bigram_words = [' '.join(bigram) for bigram, freq in common_bigrams]
frequencies = [freq for bigram, freq in common_bigrams]

plt.figure(figsize=(10, 6))
plt.barh(bigram_words[::-1], frequencies[::-1])
plt.xlabel('Frequency')
plt.title('Top 20 Bigrams')
plt.tight_layout()
plt.show()
```

```python
# === 6. Save Bigram Frequencies to CSV ===
bigram_df = pd.DataFrame(bigram_freq.items(), columns=['bigram', 'frequency'])
bigram_df['bigram'] = bigram_df['bigram'].apply(lambda x: ' '.join(x))  # convert tuples to strings

# Save to CSV
bigram_df.to_csv('bigram_frequencies.csv', index=False)
```

## Timeline Analysis
Timeline analysis was used to study how the frequency of key terms changed over time in the collected text data. Using the keywords generated from the earlier bigram analysis, this method helped track when certain words began to appear more frequently. The goal was to identify patterns in public discourse and reporting practices across the years, with the hope of understanding shifts in public attention or media framing.

This analysis was visualised in two line graph blocks.

```python
import pandas as pd
# === 1. Load CSV from existing file ===
file_path = '/content/scraped_articles_with_metadata copy.csv'  # <-- Replace with your actual filename
df = pd.read_csv(file_path)

# === 2. Define your column names ===
date_column = 'Publication Date'
text_column = 'Content'

# === 3. Clean and prepare data ===
df[date_column] = pd.to_datetime(df[date_column], errors='coerce')
df = df.dropna(subset=[date_column])
df = df.dropna(subset=[text_column])
```

Most keywords were taken from the bigram analysis while some were added according to authorial discretion.

```python
# Set time granularity (e.g., month)
df['time_period'] = df[date_column].dt.to_period('M').dt.to_timestamp()

# === 4. Define keywords to track ===
keywords = [
    '1960', 'abandonment', 'abandoned', 'act', 'animal welfare', 'arrested',
    'assault', 'attacked', 'beaten', 'beating', 'battered', 'burned',
    'burnt', 'caged', 'case', 'chained', 'charged', 'code', 'complaint',
    'cctv', 'dead', 'died', 'dismembered',
    'found', 'filed', 'footage', 'harmed', 'hospital', 'ill-treated',
    'injured', 'investigation', 'killed', 'killing', 'locked', 'maimed',
    'malnourished', 'maiming', 'media', 'mischief', 'mistreated', 'murder',
    'murdered', 'neglect', 'new delhi', 'penal', 'peta', 'place', 'poisoned',
    'police', 'report', 'right', 'suffocated', 'sexual',
    'slaughtered', 'social', 'starved', 'stabbed', 'strangled', 'stray', 'thrashed', 'tied', 'torture', 'violence'
]

# === 5. Initialize keyword columns ===
for kw in keywords:
    df[kw] = df[text_column].str.lower().str.count(kw)

# === 6. Group by time period and sum keyword counts ===
timeline_df = df.groupby('time_period')[keywords].sum().reset_index()
```

```python
# === 7. Plot ===
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 1: Check for valid keywords in the timeline_df
valid_keywords = [kw for kw in keywords if kw in timeline_df.columns]
missing_keywords = [kw for kw in keywords if kw not in timeline_df.columns]

if missing_keywords:
    print("⚠️ These keywords were not found and will be skipped:", missing_keywords)

# STEP 2: Sum mentions across time and pick top 10 for clarity
keyword_counts = timeline_df[valid_keywords].sum().sort_values(ascending=False)
top_keywords = keyword_counts.head(10).index.tolist()

# STEP 3: Plot the timeline
plt.figure(figsize=(16, 8))
palette = sns.color_palette("tab10", len(top_keywords))

for idx, kw in enumerate(top_keywords):
    plt.plot(
        timeline_df['time_period'],
        timeline_df[kw],
        label=kw,
        linewidth=2.5,
        marker='o',
        markersize=6,
        color=palette[idx]
    )

plt.title("Top 10 Keyword Mentions Over Time", fontsize=16)
plt.xlabel("Time", fontsize=14)
plt.ylabel("Mentions", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title="Keywords", fontsize=10)
plt.tight_layout()
plt.show()
```

But the individual mentions are harder to track this way, so they are plotted again separately.

```python
# Setup subplot layout
cols = 2
rows = -(-len(top_keywords) // cols)  # Ceiling division

fig, axes = plt.subplots(rows, cols, figsize=(10, 10), sharex=True)
axes = axes.flatten()

# Plot each keyword
for i, keyword in enumerate(top_keywords):
    axes[i].plot(timeline_df.index, timeline_df[keyword], marker='o', label=keyword)
    axes[i].set_title(keyword, fontsize=12)
    axes[i].set_ylabel('Mentions')
    axes[i].grid(True)
    axes[i].tick_params(axis='x', rotation=45)

# Hide extra subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Keyword Mentions Over Time (Top 10)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
# Add space between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.2)  # hspace = height, wspace = width
plt.show()
```
